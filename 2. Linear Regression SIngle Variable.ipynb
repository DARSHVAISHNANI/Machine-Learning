```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("C:\\Users\\darsh\\Machine Learning\\Dataset\\Houseprice.csv")
```

```python
df.head()
df[['Area']]
```

```python
plt.title('Scatter Plot of House-Price')
plt.xlabel('Area (Sq. ft)')
plt.ylabel('Price ($)')
plt.scatter(df.Area,df.Price,color='red',marker='+')
plt.show()
```

```python
#Calling the model
model = LinearRegression()          #df['Area'] = Input 1D dataframe
model.fit(df[['Area']],df.Price)    #df[['Area']] = Input 2D dataframe -- Sklearn take 2D array as an input
```

```python
input = 4000
y_pred = model.predict(np.array([input]).reshape(1,-1))
print(y_pred)
```

```python
print("Coefficient: " + str(model.coef_))
print("Intercept: " + str(model.intercept_))
```

```python
#using mathematics equation check the answer
print("X = " + str(input))
print("Model Predicted Y = " + str(y_pred))
Y = (model.coef_ * input) + model.intercept_
print("Mathematical Y = " + str(Y))
```

```python
# scatter plot of the predicted line
plt.scatter(df.Area,df.Price,color='red',marker='*')
plt.plot(df.Area,model.predict(df[['Area']]),color='blue')     # Instead of using this "model.predict" you can use Y_pred directly but dimrension
plt.show()
```
