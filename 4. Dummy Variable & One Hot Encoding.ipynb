There are two type of Categorical Data or Variable 
1. Nomial :-
   Where order is not important or there is no priority
   like for example: male and female
2. Odinal :-
   Where order is important or priority
   like undergraduate, postgraduate, and Ph.D

   So, for Nomial data - used method is One hot encoding
       for Odinal data - used method is Lable Encoder

For One Hot Encoding - there is this library of Dummies
it convert the each dummies value into a new columns and allocat 1 and 0

```python
import pandas as pd
df = pd.read_csv('C:/Users/darsh/Machine Learning/Dataset/homeprices - One Hot Encoding.csv')
df
```

```python
dummies = pd.get_dummies(df['town'])
print(dummies)
```

```python
# concat the data into the dataframe
df_new  = pd.concat([df,dummies],axis='columns')
df_new
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LB = LabelEncoder()
df_new['monroe township'] = LB.fit_transform(df_new['monroe township'])
df_new['robinsville'] = LB.fit_transform(df_new['robinsville'])
df_new['west windsor'] = LB.fit_transform(df_new['west windsor'])
df_new
```

Dummies Variable Trap 
So, it is like a malta colinear variable which confusion the machine learning model
so drop one columns
for example: 
if you have 4 dummy columns then drop any one and make it 3

```python
#drop the town column
#drop anyone columns like monroe township
df_final = df_new.drop(['town','monroe township'],axis=1)
df_final.head(10)
```

```python
#separate the dataset
Y = df_final['price']
X = df_final.drop(['price'],axis=1)
print(X)
print(Y)
```

```python
# deploy a linear regression trap
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```

```python
y_pred = model.predict(X)
print(y_pred)
```

```python
model.score(X,Y)
```
